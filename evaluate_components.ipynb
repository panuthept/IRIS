```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
import seaborn.objects as so
```

```python
#load file
head_path = '/ist-project/scads/sit/IRIS/eval_attn_comp.pickle'
mlp_path = '/ist-project/scads/sit/IRIS/eval_mlp_out_comp.pickle'
import pickle
with open(head_path, 'rb') as f:
    eval_head_comp = pickle.load(f)
with open(mlp_path, 'rb') as f:
    eval_mlp_comp = pickle.load(f)
```

# Harmful (GCG)

```python
data_name = 'Harmful (GCG)'
all_df = {}

for mode in ['mean', 'std']: 
    df = {}
    df['Layers'] = {}
    df['Components'] = {}
    df['scores'] = {}



    df['Layers'] = []
    df['Components'] = []
    df['scores'] = []

    for pos, score in eval_head_comp[mode][data_name].items():
        layer = pos.split('-')[0].split('L')[-1]
        head = pos.split('-')[-1]

        df['Layers'].append(int(layer))
        df['Components'].append(head)
        if type(score) == list:
            df['scores'].append(0)  
        else:
            df['scores'].append(score)  


    for pos, score in eval_mlp_comp[mode][data_name].items():
        layer = pos.split('-')[0].split('L')[-1]
        df['Layers'].append(int(layer))
        df['Components'].append('MLP')
        if type(score) == list:
            df['scores'].append(0)  
        else:
            df['scores'].append(score)  


    import pandas as pd
    all_df[mode] = pd.DataFrame.from_dict(df)
    all_df[mode] = all_df[mode].pivot(index="Components", columns="Layers", values="scores")


fig, ax =plt.subplots(1,2, figsize=(10, 5))
ax[0].title.set_text('Mean')
ax[1].title.set_text('STD')
sns.heatmap(all_df['mean'], ax=ax[0])
sns.heatmap(all_df['std'], ax=ax[1])
fig.tight_layout(pad=1.0)

plt.show()
```
